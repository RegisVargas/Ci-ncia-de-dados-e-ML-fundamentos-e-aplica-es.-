* [Introdução ao Numpy - Parte 1](https://www.kaggle.com/code/regisvargas/introdu-o-ao-numpy-parte-1)

O notebook "Introdução ao Numpy - Parte 1" ensina os fundamentos do NumPy, incluindo: criação de arrays unidimensionais (`np.array`), geração de sequências numéricas com `np.arange` e `np.linspace`, manipulação de tipos de dados (`dtype`) e uso de strings em arrays. É uma introdução prática e básica para iniciantes.
  
* [Introdução ao Numpy - Parte 2](https://www.kaggle.com/code/regisvargas/introdu-o-ao-numpy-parte-2)

O notebook "Introdução ao Numpy - Parte 2" ensina a criar arrays com `np.ones`, `np.zeros`, `np.empty` e `np.random.rand`, além de matrizes multidimensionais com `np.array` e `np.reshape`. Também aborda as propriedades `size` (número de elementos), `shape` (dimensões) e `ndim` (número de dimensões).
  
* [Introdução ao Numpy - Parte 3](https://www.kaggle.com/code/regisvargas/introdu-o-ao-numpy-parte-3)

O notebook "Introdução ao Numpy - Parte 3" explora operações matemáticas e indexação no NumPy. As operações aritméticas (+, -, *, /) são aplicadas elemento a elemento entre arrays, escalares também podem ser usados para alterar todos os elementos de um array. Também ensina indexação para acessar elementos específicos e slicing para extrair subconjuntos de arrays, usando a sintaxe `[início:fim:passo]`. Em arrays 2D, demonstra como selecionar linhas, colunas ou regiões específicas, tratando-os como "arrays de arrays". É útil para quem quer aprender a manipular dados de forma eficiente.

* [Introdução ao Numpy - Parte 4](https://www.kaggle.com/code/regisvargas/introdu-o-ao-numpy-parte-4)

O notebook "Introdução ao Numpy - Parte 4" explora o empilhamento e subdivisão de arrays no NumPy. Utiliza `np.vstack` (empilhamento vertical), `np.hstack` (horizontal) e `np.split` (divisão em partes iguais) para manipulação eficiente de dados.
  
* [Resolução de um sistema linear a 2 incógnitas - Parte 1](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-1)

O notebook "Resolução de um sistema linear a 2 incógnitas - Parte 1" ensina a resolver sistemas lineares com duas incógnitas usando `np.linalg.solve` do NumPy. Ele define coeficientes e resultados como arrays, resolve o sistema e verifica as soluções substituindo-as nas equações.

* [Resolução de um sistema linear a 2 incógnitas - Parte 2](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-2)

O notebook "Resolução de um sistema linear a 2 incógnitas - Parte 2" aborda a resolução de sistemas lineares com duas incógnitas, destacando:  
1. **Cálculo do determinante**: Verifica a existência de solução única usando `np.linalg.det`.  
2. **Método da eliminação**: Resolve o sistema manualmente, mostrando passos algébricos.  

É um complemento ao estudo do uso de métodos automáticos.
