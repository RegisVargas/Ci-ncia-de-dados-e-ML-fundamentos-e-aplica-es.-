* [Resolução de um sistema linear a 2 incógnitas - Parte 5](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-5)

O notebook "Resolução de um sistema linear a 2 incógnitas - Parte 5" analisa sistemas lineares com determinante da matriz de coeficientes igual a zero, focando em casos com **infinitas soluções**. O sistema apresentado $x + 2y = 3$ e $2x + 4y = 6$ é redundante, pois a segunda equação é o dobro da primeira. Isso reduz o sistema a $x + 2y = 3$, cuja solução geral é $y = -\frac{1}{2}x + \frac{3}{2}$. Assim, qualquer valor de $x$ gera uma solução válida, como $(1, 1)$.

* [Resolução de um sistema linear a 2 incógnitas - Parte 6](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-6)

O notebook "Resolução de um sistema linear a 2 incógnitas - Parte 6" apresenta um sistema linear sem solução, demonstrando o caso em que a matriz de coeficientes tem determinante igual a zero e as equações são inconsistentes.

### Sistema analisado:
$$
\begin{cases}
1x + 1y = 2 \\
2x + 2y = 8
\end{cases}
$$

### Etapas principais:
1. **Matriz de coeficientes e resultados**:
   - Matriz de coeficientes: $\begin{bmatrix} 1 & 1 \\ 2 & 2 \end{bmatrix}$
   - Determinante da matriz: $0$, indicando um sistema singular.

2. **Tentativa de resolução**:
   - Uso de `np.linalg.solve` retorna erro de "Singular matrix".

3. **Verificação algébrica**:
   - Substituindo $x = 2 - y$ da primeira equação na segunda resulta na contradição $4 = 8$, comprovando que o sistema é inconsistente.

4. **Representação gráfica**:
   - As equações são transformadas em retas:
     - $r: y = -x + 2$
     - $s: y = -x + 4$
     - As retas são paralelas, sem interseção, confirmando a inexistência de solução.
