O pacote Numpy é utilizado para criação de arrays e matrizes. Além disso, facilita o cálculo de operações matemáticas. Vamos começar importando este pacote.

```python
import numpy as np
```

O tipo de dado para o array Numpy é o *ndarray*. O array mais simples do numpy é o array de uma dimensão, que pode, por exemplo, ser definido como no código abaixo.

```python
array_unidimensional = np.array([10, 12, 14])
print(array_unidimensional)
```

Para criar uma listagem de números igualmente espaçados, utiliza-se *np.arange*. O código abaixo gera um lista de números inteiros de zero a 8. Observe que **o limite superior (9) não é incluído na listagem**.

```python
np.arange(9)
```

Já, para criar uma lista com números limitados entre, por exemplo, 1 e 13, variando de 2 em 2. Fazemos:

```python
np.arange(1,14,2)
```

Observe novamente **que o limite superior (14) não é incluído na listagem**. Por essa razão o código abaixo mantém a mesma saída do código acima.

```python
np.arange(1,15,2)
```

A função *np.linspace* também é usada para criar lista de números igualmente espaçados. Porém há uma diferença entra a função *np.linspace* e a função *np.arange*. Enquanto o *np.arange* define "passos" entre os números, o *np.linspace* define quantos números igualmente espaçados devem ser criados em um intervalo pré-determinado. Por exemplo: criar 6 valores (igualmente espaçados) entre 1 e 15.

```python
np.linspace(1, 15, 6)
```

Outras diferenças a se considerar:

1. Os extremos agora são aceitos na listagem
2. Os dados, agora são do tipo float.

Você pode gerar a lista anterior somente com números inteiros. Para isto, basta usar o *dtype = int*:

```python
np.linspace(1, 15, 6, dtype = int)
```

Vamos ver, agora, como o Numpy lida com *strings*. Considere a seguinte frase "estudando ciência de dados pelos fundamentos". Essa frase, para um array numpy é considerada como um único elemento. Então, ele vai simplesmente "printar" a frase. Veja:

```python
np.array("estudando ciência de dados pelos fundamentos")
```

*dtype='<U44'* indica que a frase tem 44 caracteres *unicode*. Isto, contados os espaços em branco.
