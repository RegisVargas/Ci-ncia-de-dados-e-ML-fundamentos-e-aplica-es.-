```python
import numpy as np
```

A função *np.ones* cria um array somente de uns. Sendo necessário apenas passar o tamanho do array.

```python
print(np.ones(10))
```

De forma análoga, temos as funções:

* *np.zeros* - retorna um array de zeros
* *np.empty* - retorna um array não inicializado
* *np.random.rand* - retorna um array de números aleatórios 

```python
print(np.zeros(10))
```

```python
print(np.empty(10))
```

```python
print(np.random.rand(3))
```

O Numpy também permite a criação de arrays multidimensionais. Como, por exemplo, uma matriz de 3 linhas por 2 colunas. Veja abaixo.

```python
matriz_3_2 = np.array([
    [1,2], 
    [4,5],
    [2,3]
]
                     )
print(matriz_3_2)
```

Essa matriz também pode ser criada utilizando o *np.reshape*

```python
elementos = np.array([1,2,4,5,2,3])
print(elementos.reshape(3,2))
```

Encontrando o tamanho de uma matriz (neste caso, 6 elementos)...

```python
matriz_3_2.size
```

Sua forma (neste caso, 3 linhas por 2 colunas)... 

```python
matriz_3_2.shape
```

E sua dimensão. Neste caso, a dimensão é 2. Uma indicando a linha e a outra indicando a coluna.

```python
matriz_3_2.ndim
```
