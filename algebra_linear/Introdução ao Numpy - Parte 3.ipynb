```python
import numpy as np
```

Neste notebook, veremos como funcionam as operações matemáticas com arrays numpy. Vamos começar criando os vetores de exemplo *a* e *b*.

```python
a = np.array([4, 8, 9])
b = np.array([8, 2, 3])
```

As operações aritméticas são aplicadas aos elementos correspondentes. Veja abaixo.

```python
# [4+8, 8+2, 9+3]
print(a+b)
```

```python
# [4-8, 8-2, 9-3]
print(a-b)
```

```python
# [4×8, 8×2, 9×3]
print(a*b)
```

```python
#[4÷8, 8÷2, 9÷3]
print(a/b)
```


Ao multiplicar um escalar por um array (e vice-versa), o resultado será um array com o mesmo formato (shape), onde todos os elementos do array original são multiplicados pelo escalar. Veja abaixo um exemplo para o array *a* e o escalar *1.5*

```python
#[4×1.5, 8×1.5, 9×1.5]
print(1.5*a)
```

O mesmo vale para as outras operações aritméticas.

```python
#[4÷2, 8÷2, 9÷2]
print(a/2)
```

```python
#[4+3, 8+3, 9+3]
print(a+3)
```

```python
#[4-5, 8-5, 9-5]
print(a-5)
```

O primeiro índice de um array é o *zero. Assim: *a[0], a[1] e a[2]* valem, respectivamente, *4, 8, e 9*

```python
print(a[0], a[1], a[2])
```

Vamos ver agora como obter apenas uma parte do array. A sintaxe para isso é *array[inicio:passo:fim]*. 

* Se não for passado o *inicio*, então, inicio será considerado zero;

* Se não for passado *passo*, então *passo* será considerado 1;

* Se não for passado *fim*, então *fim* será considerado *n-1* onde *n* é a quantidade de elementos no array. Lembre-se que o índice do array começa em zero.

Veja os exemplos abaixo.

```python
ex = np.arange(2,8)
print(ex)
```

No exemplo abaixo, temos *inicio=0*, *passo=1* e *fim=1*.

Então, é retornado somente o primeiro elemento do array *ex*.

Isto ocorre porque o primeiro elemento a ser listado tem índice igual a *inicio* e o último elemento a ser listado tem índice igual a *fim-1*.

Assim, tanto o primeiro quanto o último elemento da lista são ex[0].

```python
print(ex[:1])
```

Abaixo, temos *inicio=1*, *passo=1* e *fim=4*. Então, serão retornados os elementos de índices 1, 2 e 3.

```python
print(ex[1:4])
```

Podemos retornar os dados do array de 2 em 2, por exemplo. Informando só o *passo*.

```python
print(ex[::2])
```

Vamos ver agora, como esta estrutura funciona em um array de 2 dimensões. Veja abaixo que, neste caso, *ex_2d* é um "array de arrays". Isto é, o elemento no de índice zero é \[3,7,8]. Já o elemento de índice 1 é o \[4,5,8] e, por fim, o elemento de índice 2 é o \[2,4,7].

```python
ex_2d = np.array([
    [3,7,8],
    [4,5,8],
    [2,4,7]
])
```

Então, o código abaixo, por exemplo, deve retornar os dois primeiros elementos de *ex_2d*.

```python
print(ex_2d[0:2])
```

Se quisermos, por exemplo, os dois primeiros números da primeira coluna, fazemos:

```python
print(ex_2d[0:2,0])
```
