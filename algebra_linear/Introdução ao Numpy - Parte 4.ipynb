```python
import numpy as np
```

Neste notebook veremos o recurso de empilhamento e subdivisão do Numpy. Vamos começar definindo os arrays *a* e *b*.

```python
a = np.array([
    [1,7,5],
    [2,4,8],
    [1,9,3]
])

b = np.array([
    [2,4,5],
    [1,7,3],
    [4,9,8]
])
```

Estes dois arrays podem ser empilhados verticalmente usando a função *np.vstack*. Veja abaixo.

Observação: note a necessidade em se passar os arrays a serem empilhados, entre parenteses: *(a,b)*. Se você fizer *np.vstack(a,b)* ao invés de *np.vstack(**(**a,b**)**)* resultará em erro.

```python
print(np.vstack((a,b)))
```

De maneira análoga, podemos fazer o empilhamento horizontal com a função *np.hstack*. Veja abaixo.

```python
print(np.hstack((a,b)))
```

Temos também a função *np.split*, que nos permite subdividir um array. Abaixo, fazemos o empilhamento vertical de *a* e *b* e depois subdivimos o array empilhado em 3 arrays de mesmo tamanho.

```python
a_vstack_b = np.vstack((a,b))

split_a_vstack_b = np.split(a_vstack_b,3)

print("array 1 \n",split_a_vstack_b[0])

print("array 2 \n",split_a_vstack_b[1])

print("array 3 \n", split_a_vstack_b[2])
```
