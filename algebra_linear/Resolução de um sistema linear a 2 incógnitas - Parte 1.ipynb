<h1>Resolução de um sistema linear a 2 incógnitas - Parte 1</h1>

Suponha que você precise encontrar dois números, $x$ e $y$ tais que:

$$
\begin{cases}
-x+y=1\\
4x-2y=4
\end{cases}
$$

Sabemos que quando não aparece número na frente da letra, é porque na frente da letra tem o número 1. Então, podemos reescrevar o sistema anterior da seguinte forma:

$$
\begin{cases}
-1x+1y=1\\
4x-2y=4
\end{cases}
$$

Podemos, utilizando *np.array*, criar a matriz dos coeficientes deste sistema de equações, isto é os números que acompanha as letras. Veja abaixo como fazer isto.

```python
import numpy as np

coeficientes = np.array([
    [-1,1],
    [4,-2]
])
```

Você também pode criar a matriz com os resultados de cada equação usando um raciocínio semelhante. Veja abaixo.

```python
resultados = np.array([
    [1],
    [4]
])
```

O *numpy* promete que consegue resolver esta equação rapidinho através da função *np.linalg.solve*. Vamos verificar se esta não é uma promessa vazia. 

Para aplicar a função *np.linalg.solve* basta passar como parâmetros a matriz de coeficientes e a matriz dos resultados.

```python
print(np.linalg.solve(coeficientes, resultados))
```

A função *np.linalg.solve* nos diz que as soluções para o sistema de equações apresentado são $x=3$ e $y=4$.

Vamos fazer a prova real e verificar se, de fato, estas são soluções das equações que compõem o sistema de equações apresentado.

A primeira equação é $-x+y=1$. De fato, trocando $x$ por $3$ e $y$ por $4$, a igualdade é satisfeita, veja abaixo:

$-3+4=1$

A segunda equação é $4x-2y=4$. Novamente, trocando $x$ por $3$ e $y$ por $4$. Temos:

$4\cdot 3-2\cdot 4 = 12-8=4$

Vemos, assim, que a função *np.linalg.solve* resolve o sistema de equações fornecido de forma correta.
