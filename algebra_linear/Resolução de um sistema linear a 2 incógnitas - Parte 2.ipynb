Abaixo, temos o sistema de equações apresentado no [notebook anterior](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-1)

$$
\begin{cases}
-1x+1y=1\\
4x-2y=4
\end{cases}
$$

Vimos também que a matriz de coeficientes para este sistema é:

$$
\begin{bmatrix}
-1 & 1 \\
4 & -2
\end{bmatrix}
$$

Além disso, sabemos que as soluções deste sistema são $x=3$ e $y=4$. Poderíamos, antecipadamente, saber que este sistema possui **uma única solução**, calculando o **determinate** da matriz dos coeficientes, que é **diferente de zero**.

Sabemos que, no caso da matriz dois por dois, o determinante é calculado da seguinte forma:

(produto dos elementos da diagonal principal) - (produto dos elementos da diagonal secundária)


Isto é:

$(-1)\times (-2) -(1\times 4)=2-4=-2≠0$

Como se sabe, calcular o determinante de uma matriz pode ser muito difícil (a depender do tamanho desta matriz). 

Para facilitar este processo, podemos utilizar a função *np.linalg.det*. Observe que esta função encontra o mesmo valor de $-2$ para o determinante da matriz dos coeficientes.

```python
import numpy as np

coeficientes = np.array([
    [-1,1],
    [4,-2]
])

print(
    np.linalg.det(
        coeficientes
    )
)
```

Até o momento, apenas utizamos, no [notebook anterior](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-1), a função *np.linalg.solve* e testamos as soluções encontradas por esta função no sistema de equações. Isto é, fizemos a prova real.

Contudo, é possível resolver este sistema manualmente através, por exemplo do método da eliminação.

Retornando ao sistema de equações dado por:

$$
\begin{cases}
-x+y=1\\
4x-2y=4
\end{cases}
$$

Podemos, por exemplo, multiplicar a primeira equação por $4$, o que resultaria no sistema (equivalente) dado abaixo:

$$
\begin{cases}
-4x+4y=4\\
4x-2y=4
\end{cases}
$$

Somando ambos os lados da igualdade das equações acima temos:

$$
-4x+4x+4y-2y=4+4
$$
$$
2y=8
$$
$$
y=\frac{8}{2}
$$
$$
y=4
$$

Substituindo $y=4$ na segunda equação, por exemplo, temos que:

$$
4x-2\cdot 4=4
$$
$$
4x-8=4
$$
$$
4x=4+8
$$
$$
4x=12
$$
$$
x=\frac{12}{4}
$$
$$
x=3
$$
