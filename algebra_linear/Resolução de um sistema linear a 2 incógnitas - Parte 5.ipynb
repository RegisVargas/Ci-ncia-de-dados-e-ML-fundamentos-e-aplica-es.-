[Já vimos](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-2?scriptVersionId=218363306&cellId=6) que quando o **determinante** da matriz de coeficientes de um sistema de equações lineares é **diferente de zero**, o sistema apresenta **uma única solução**.

Neste notebook, abordaremos um dos casos em que este determinante é igual a zero. Para tal, trataremos de um sistema de equações que apresenta infinitas soluções. Veja abaixo.

$$
\begin{cases}
1x+2y=3\\
2x+4y=6
\end{cases}
$$

A matriz de coeficientes deste sistema é dada por

```python
import numpy as np
coeficientes=np.array([
[1,2],
[2,4]])
print(coeficientes)
```

O cálculo do determinante desta matriz é igual a zero. Veja abaixo.

```python
print(np.linalg.det(coeficientes))
```

Quando o determinante da matriz de coeficientes é zero, existem duas possibilidades:

1. O sistema não tem solução;
2. O sistema possui infinitas soluções.

Vamos, então, analisar o sistema de equações apresentado no início deste notebook.

$$
\begin{cases}
1x+2y=3\\
2x+4y=6
\end{cases}
$$

Se você observar bem, verá que a segunda equação pode ser obtida a partir da primeira, multiplicando-a (em ambos os lados de sua igualdade) por $2$. Veja:

$$
1x+2y=3\times (2)
$$
$$
2x+4y=6
$$

Assim, podemos dizer que o sistema de equações é redundante e equivale a apresentar apenas a equação abaixo:

$$
1x+2y=3
$$

Isolando o $y$ nesta equação, temos:

$$
y=-\frac{1}{2}x+\frac{3}{2}
$$

Assim, escolhendo um valor para $x$, a partir da equação acima, formamos um par par $(x,y)$ que satisfaz o sistema apresentado. Por exemplo, escolhendo $x=1$ o par $(1,1)$ satisfaz o sistema apresentado. Veja:

$$
\begin{cases}
1\cdot 1+2\cdot 1=3\\
2\cdot 1+4\cdot 1=6
\end{cases}
$$

Como podemos escolher infinitos valores para $x$, então podemos formar infinitas soluções para o sistema apresentado neste notebook.
