Neste notebook, veremos um exemplo de sistema de equações cujo determinante da matriz de coeficientes é igual a zero. 

[Como já vimos](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-5), no caso deste determinante ser igual a **zero**, o sistema apresentará **infinitas soluções** ou **nenhuma solução**.

O exemplo que será apresentado neste notebook é um tipo de **sistema sem solução**.

Considere o sistema abaixo.

$$
\begin{cases}
1x+1y=2 \\
2x+2y=8
\end{cases}
$$

[Como já vimos,](https://www.kaggle.com/code/regisvargas/sistema-linear-parte-1) podemos montar a matriz de coeficientes da seguinte forma:

```python
import numpy as np

coeficientes = np.array([
    [1,1],
    [2,2]
])

print(coeficientes)
```

Também podemos montar a matriz de resultados.

```python
resultados = np.array([
    [2],
    [8]])
print(resultados)
```

Fazendo o empilhamento horizontal, obtemos a matriz que representa o sistema.

```python
sistema = np.hstack((
    coeficientes,
    resultados
))
print(sistema)
```

Note que, ao calcular o determinante da matriz de coeficientes, obtemos o valor zero.

```python
print(np.linalg.det(coeficientes))
```

E, ao tentar resolver o sistema, recebemos o erro *Singular matrix*.

```python
print(np.linalg.solve(coeficientes, resultados))
```

Isto é, o determinante da matriz de coeficientes é igual a zero. Vamos verificar que, para o caso deste exemplo. O sistema não tem solução. Para isso, começamos isolando o $x$ na primeira equação.

$$
x+y=2
$$
$$
x=2-y
$$

Agora, vamos substituir o valor para $x$, encontrado acima, na segunda equação.

$$
2\cdot (2-y) + 2y = 8
$$
$$
4-2y+2y=8
$$
$$
4=8
$$

Como $4=8$ é uma igualdade falsa, temos que o sistema apresentado não tem solução.

Para verificarmos isso de forma gráfica, isolamos o $y$ nas duas equações. Obtendo, assim, as equações da reta $r$ e $s$, dadas abaixo.

$$
r:y=-x+2
$$

$$
s:y=-x+4
$$

Podemos então, plotar as retas $r$ e $s$ no mesmo plano cartesiano. Veja abaixo.

```python
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

yr = -x+2

ys = -x+4

plt.grid(True);
plt.plot(x, yr, label='$r:y = -x + 2$', color='blue', linewidth=2);      # Reta r
plt.plot(x, ys, label='$s:y = -x + 4$', color='orange', linewidth=2);    # Reta s
plt.xlabel('$x$', fontsize=14);
plt.ylabel('$y$', fontsize=14);
plt.legend(fontsize=12);
```

Vemos então duas retas paralelas representando o nosso sistema de equações. Como, por definição, retas paralelas não têm pontos em comum, isso equivale a dizer que o sistema não tem solução.

Era de se esperar um gráfico assim. Já que as duas retas possuem o mesmo coeficiente angular, de valor $-1$, e coeficientes lineares de valores, respectivamente $2$ e $4$. Na prática, a reta $s$ equivale a reta $r$ deslocada $2$ unidades verticalmente, para cima.
